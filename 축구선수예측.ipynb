# Data Preparation
## train, validation data 만들기

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=['Prospect'])
y = df.Prospect

# 트테트테
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.3, random_state = 42) # 42는 우주의 신비

print(X_train.shape, X_val.shape, y_train.shape, y_val.shape)
```

# Training

```python
# 1.SVM
from sklearn.svm import SVC
# 2. RandomForest
from sklearn.ensemble import RandomForestClassifier
# 3. lda
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
# 4. Naive
from sklearn.naive_bayes import GaussianNB

clf = SVC(C=0.2, degree=2, kernel='linear')
clf2 = RandomForestClassifier(n_estimators=80, max_depth=8)
lda = LDA(n_components=1,tol=0.0002)
gnb = GaussianNB()

clf.fit(X_train, y_train) #학습
clf2.fit(X_train, y_train)
lda.fit(X_train, y_train)
gnb.fit(X_train, y_train)
```

# Evaluation

```python
from sklearn.metrics import f1_score

pred_train = clf.predict(X_train)
pred_val = clf.predict(X_val)

pred_train2 = clf2.predict(X_train)
pred_val2 = clf2.predict(X_val)

pred_train3 = lda.predict(X_train)
pred_val3 = lda.predict(X_val)

pred_train4 = gnb.predict(X_train)
pred_val4 = gnb.predict(X_val)

print("---SVM---")
print("Train Accuracy : %.4f" % f1_score(y_train, pred_train))
print("Validation Accuracy : %.4f" % f1_score(y_val, pred_val))

print("---Random Forest---")
print("Train Accuracy : %.4f" % f1_score(y_train, pred_train2))
print("Validation Accuracy : %.4f" % f1_score(y_val, pred_val2))

print("---LDA---")
print("Train Accuracy : %.4f" % f1_score(y_train, pred_train3))
print("Validation Accuracy : %.4f" % f1_score(y_val, pred_val3))

print("---GNB---")
print("Train Accuracy : %.4f" % f1_score(y_train, pred_train4))
print("Validation Accuracy : %.4f" % f1_score(y_val, pred_val4))
```
